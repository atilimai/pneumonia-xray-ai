```python
import os
import matplotlib.pyplot as plt

base_path = "data/raw/chest_xray"

splits = ["train", "val", "test"]
classes = ["NORMAL", "PNEUMONIA"]

data = {}

for split in splits:
    data[split] = []
    for cls in classes:
        path = os.path.join(base_path, split, cls)
        count = len(os.listdir(path))
        data[split].append(count)

# Plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, split in enumerate(splits):
    axes[i].bar(classes, data[split])
    axes[i].set_title(split)
    axes[i].set_ylabel("Image Count")

plt.tight_layout()

# Save
os.makedirs("artifacts/figures", exist_ok=True)
plt.savefig("artifacts/figures/class_distribution.png")

plt.show()
```
